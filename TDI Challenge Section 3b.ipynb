```python
import pandas as pd
import scipy as sp
from scipy import stats
import math
import pylab
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('NYPD_Arrests_Data__Historic_.csv')
df.head()
```

```python
df = df.dropna()
```

```python
df['PD_DESC'].value_counts()
```

```python
df2 = df[df['PD_DESC'] == 'MARIJUANA, POSSESSION 4 & 5']
```

```python
df3 = pd.DataFrame(pd.to_datetime(df2['ARREST_DATE']).dt.year.value_counts())
df3 = df3.reset_index()
df3.columns = ['Year', 'Count']
df3
```

```python
plt.bar(df3['Year'],df3['Count'])
plt.title('Marijuana Possesion Crime Count vs Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```
